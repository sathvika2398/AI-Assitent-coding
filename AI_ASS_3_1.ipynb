## 1. Zero-Shot Prompting (Palindrome Number Program)

“Write a Python function to check whether a given number is a palindrome.”

AI Generated code:

```python
def is_palindrome(n):
    s = str(n)
    return s == s[::-1]

n = int(input())
print("Palindrome" if is_palindrome(n) else "Not Palindrome")
```

### Testing outputs:

232
palindrome

123
Not Palindrome

observation: The program checks whether the given number reads the same forwards and backwards.

## 2. One-Shot Prompting (Factorial Calculation)


“Write a Python function to calculate factorial of a number.
Example: Input: 5 → Output: 120.”

AI Generated code:

```python
def factorial(n):
    if n < 0:
        return None
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

n = int(input())
print(factorial(n))
```

### output:

5

120

observation: The program calculates and prints the factorial of a given non-negative number.

## 3. Few-Shot Prompting (Armstrong Number Check)

“Write a Python function to check whether a number is an Armstrong number.
Examples: 153 → Armstrong, 123 → Not Armstrong.”

AI Generated code:

```python
def is_armstrong(n):
    digits = str(n)
    power = len(digits)
    total = sum(int(d)**power for d in digits)
    return total == n
n = int(input())
print("Armstrong Number" if is_armstrong(n) else "Not an Armstrong Number")
```

### Testing outputs:

153->Amstrong Number

123->Not an Amstrong Number

Observation:
The program checks whether the given number is an Armstrong number by comparing the sum of its digits raised to the number of digits with the original number.

## 4. Context-Managed Prompting (Optimized Number
## Classification)


### “Write an optimized Python program to classify a number as Prime, Composite, or Neither with proper input validation.”

AI Generated code:

```python
def classify(n):
    if n <= 1:
        return "Neither"
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return "Composite"
    return "Prime"
n = int(input())
print(classify(n))
```

### Testing outputs:

13->Prime

34->Composite

observation:By the above outputs 13 will not divisible by any number except with itself and 1 so it is prime and 34 is even number which is divible so it is composite number

## 5. Zero-Shot Prompting (Perfect Number Check)

“Write a Python function to check whether a given number is a perfect number.”

AI Generated code:

```python
def is_perfect(n):
    if n <= 0:
        return False
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total == n
n = int(input())
print("Perfect Number" if is_perfect(n) else "Not a Perfect Number")
```

### Testing outputs:

6

Perfect Number

10

Not a Perfect Number

Observation:
When the input is 6, the output is “Perfect Number”, and when the input is 10, the output is “Not a Perfect Number” because only 6 equals the sum of its proper divisors.
